# Introduction

This analysis explores how market sentiment (Fear vs Greed) impacts trader behavior and performance on Hyperliquid. The goal is to identify patterns that can inform better trading strategies.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sentiment = pd.read_csv("fear_greed_index.csv")
trades = pd.read_csv("historical_data.csv")
```

### Data Overview
- Dataset contains X trades across Y accounts.
- Missing values are minimal / handled.
- No significant duplicates affecting analysis.

```python
print("Sentiment shape:", sentiment.shape)
print("Trades shape:", trades.shape)

print("\nMissing values:\n", trades.isnull().sum())
print("\nDuplicates:", trades.duplicated().sum())
```

```python
sentiment.shape
trades.shape

sentiment.info()
trades.info()

sentiment.isnull().sum()
trades.isnull().sum()

trades.duplicated().sum()
```

```python
sentiment['date'] = pd.to_datetime(sentiment['date'])
trades['Timestamp IST'] = pd.to_datetime(trades['Timestamp IST'], format='%d-%m-%Y %H:%M')
```

```python
trades['date'] = trades['Timestamp IST'].dt.date
sentiment['date_only'] = sentiment['date'].dt.date
```

### Date Alignment
- Converted timestamps to daily granularity for merging datasets.

```python
trades['date'] = trades['Timestamp IST'].dt.date
sentiment['date'] = pd.to_datetime(sentiment['date']).dt.date
```

### Data Merge
- Merged trader data with sentiment using date.
- Standardized column names for consistency.

```python
df = trades.merge(sentiment, on='date', how='left')
```

```python
df.columns = df.columns.str.lower()
```

```python
daily_pnl = df.groupby(['Account','date_x'])['Closed PnL'].sum().reset_index()
```

```python
df['win'] = df['Closed PnL'] > 0
win_rate = df.groupby('Account')['win'].mean()
```

```python
trades_per_day = df.groupby('date_x').size()
```

### Feature Engineering
- Created win indicator (profitability per trade)
- Constructed leverage proxy using normalized position size

```python
df['abs_size'] = df['size usd'].abs()
df['avg_size_per_trader'] = df.groupby('account')['abs_size'].transform('mean')
df['leverage_proxy'] = df['abs_size'] / df['avg_size_per_trader']
```

## Trader Segmentation

```python
df['lev_group'] = np.where(df['leverage_proxy'] > df['leverage_proxy'].median(), 'High', 'Low')
```

```python
# Re-calculating the segments using correct lowercase column names
df['abs_size'] = df['size usd'].abs()
df['avg_size_per_trader'] = df.groupby('account')['abs_size'].transform('mean')
df['leverage_proxy'] = df['abs_size'] / df['avg_size_per_trader']
df['lev_group'] = np.where(df['leverage_proxy'] > df['leverage_proxy'].median(), 'High', 'Low')

# Grouping and calculating mean PnL using standardized column names
segment_analysis = df.groupby(['lev_group', 'classification'])['closed pnl'].mean().reset_index()
print(segment_analysis)
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='lev_group', y='closed pnl', data=df)
plt.title("PnL by Trader Exposure")
plt.show()
```

```python
freq = df.groupby('Account').size()
df['freq_group'] = df['Account'].map(lambda x: 'High' if freq[x] > freq.median() else 'Low')
```

```python
# Re-calculating frequency groups to ensure the column exists
freq = df.groupby('account').size()
median_freq = freq.median()
df['freq_group'] = df['account'].map(lambda x: 'High' if freq[x] > median_freq else 'Low')

# Performing the analysis
print(df.groupby('freq_group')['closed pnl'].mean())
```

**Insight:**  
High exposure traders face larger losses during Fear periods, while frequent traders tend to overtrade and reduce profitability.

```python
df.groupby(['lev_group','classification'])['closed pnl'].mean()
```

```python
pnl = df.groupby('Account')['Closed PnL'].sum()
df['profit_group'] = df['Account'].map(lambda x: 'Winner' if pnl[x] > 0 else 'Loser')
```

```python
df.groupby('classification')['leverage_proxy'].mean()
df.groupby('classification').size()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='classification', y='leverage_proxy', data=df)
plt.title("Trader Aggressiveness by Sentiment")
plt.show()
```

**Insight:**  
Traders tend to take larger positions during Greed periods, reflecting increased risk appetite.

# Key Insights

1. Market Fear leads to higher volatility and lower average trader profitability.
2. High exposure traders are significantly more vulnerable during Fear periods.
3. Frequent traders tend to overtrade during Greed phases, reducing net returns.

# Strategy Recommendations

1. During Fear periods, traders should reduce position sizes and avoid high exposure trades, as volatility leads to disproportionately larger losses for aggressive traders.
2. During Greed phases, traders should limit trade frequency, as increased activity correlates with overtrading and reduced profitability.
